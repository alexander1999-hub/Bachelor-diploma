```python
import numpy
from matplotlib import pyplot as plt
import numpy as np
import os
```

```python
path = os.path.dirname(os.path.abspath('TXT_TO_PNG.ipynb'))
filelist = os.listdir(path)
counter = 0
for i in filelist:
    if i.endswith(".txt"):  
        with open(path + '\\' + i, 'r') as f :           
            #f = open("aim0-0.txt", "r")
            a = f.read().split('\n')[:-1]
            a1_1 = []
            for line in a :
                a1 = line.split(' ')
                if '' in a1 :
                    a1.remove('')
                a1_1.append(list(map(float, a1)))
            a1_2 = np.stack([elem for elem in a1_1])

            x = a1_2 * 255
            fig = plt.figure(frameon=False)
            DPI = fig.get_dpi()
            fig.set_size_inches(300.0/float(DPI),300.0/float(DPI))
            ax = plt.Axes(fig, [0., 0., 1., 1.])
            ax.set_axis_off()
            fig.add_axes(ax)
            ax.imshow(x, aspect='auto', cmap='gray', origin='lower')
            fig.savefig(i[:-4] + '.png')
            counter+=1
            plt.close()
```
